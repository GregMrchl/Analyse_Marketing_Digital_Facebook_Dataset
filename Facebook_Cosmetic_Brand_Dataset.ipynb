## Nettoyage des données

```python
# Est-ce qu'il y a des valeurs nulles ?
df.isnull().sum()
```

```python
# Est-ce qu'il y a des doublons ?
df.duplicated().sum()
```

```python
# Gérer les valeurs manquantes avant de convertir en int64.
# Utilisation de la médiane pour l'imputation afin d'éviter les erreurs avec astype(int).

df['like'] = df['like'].fillna(df['like'].median())
df['share'] = df['share'].fillna(df['share'].median())

# Changer le type des colonnes 'like' et 'share' en int64
df['like'] = df['like'].astype('int64')
df['share'] = df['share'].astype('int64')

# La colonne 'Category' n'est pas pertinente car il n'y a aucune information écrite dans la description du dataset.
# Suppression de la colonne 'Category', ignorer l'erreur si elle n'existe pas.
df.drop('Category', axis=1, inplace=True, errors='ignore')

print("DataFrame après modifications:")
display(df.info())
display(df.head())
```

##Début de l'analyse

```python
# Calcul du taux d'engagement.
# Utilisation des colonnes 'Total Interactions' et 'Lifetime Post Total Reach' pour le calcul.
# Il est important de gérer les cas où 'Lifetime Post Total Reach' pourrait être égal à zéro pour éviter les erreurs de division.
df['Engagement Rate'] = np.where(
    df['Lifetime Post Total Reach'] > 0,
    ((df['Total Interactions'] / df['Lifetime Post Total Reach']) * 100).round(2),
    0
)

# Affichage et mise en pourcentage de la nouvelle colonne 'Engagement Rate'.
display(df[['Total Interactions', 'Lifetime Post Total Reach', 'Engagement Rate']].head().style.format({'Engagement Rate': '{:.2f}%'}))
```

```python
# 1. Vérifier s'il y a des valeurs Inf
print(f"Valeurs Inf : {df['Engagement Rate'].isin([np.inf, -np.inf]).sum()}")

# 2. Vérifier s'il y a des NaN
print(f"Valeurs NaN : {df['Engagement Rate'].isna().sum()}")

# 3. Vérifier le type de données
print(f"Type : {df['Engagement Rate'].dtype}")

# 4. Regarder quelques valeurs
print(df[['Total Interactions', 'Lifetime Post Total Reach', 'Engagement Rate']].head(20))

# 5. Stats descriptives
print(df['Engagement Rate'].describe())
```

Cette méthode nous donne un taux d'engagement direct, indiquant le pourcentage de personnes uniques ayant vu la publication et ayant interagi avec elle. C'est une métrique clé pour évaluer la performance des publications.

```python
# Répartition par type de post
print("Répartition des posts par type :\n")
display(df['Type'].value_counts())
```

```python
# Taux d'engagement moyen par type de post
engagement_par_type = df.groupby('Type')['Engagement Rate'].mean().sort_values(ascending=False).round(2)

print("\nTaux d'engagement moyen par type de post :\n")
display(engagement_par_type)

# Type de post avec le taux d'engagement le plus élevé
type_plus_engageant = engagement_par_type.index[0]
taux_plus_engageant = engagement_par_type.iloc[0]

print(f"\nLe type de post avec le taux d'engagement le plus élevé est '{type_plus_engageant}' avec un taux moyen de {taux_plus_engageant:.2f}%.")
```

```python
# Taux d'engagement moyen par heure de publication
engagement_par_heure = df.groupby('Post Hour')['Engagement Rate'].mean().sort_values(ascending=False).round(2)

print("\nTaux d'engagement moyen par heure de publication :\n")
display(engagement_par_heure)

# Heure de publication avec le taux d'engagement le plus élevé
heure_plus_engageante = engagement_par_heure.index[0]
taux_plus_engageant_heure = engagement_par_heure.iloc[0]

print(f"\nL'heure de publication qui génère le plus d'engagement est {heure_plus_engageante}h avec un taux moyen de {taux_plus_engageant_heure:.2f}%.")
```

```python
# Filtrer les posts de type 'Photo' publiés à 23h
posts_photo_23h = df[(df['Type'] == 'Photo') & (df['Post Hour'] == 23)]

# Compter le nombre de ces posts
nombre_posts_photo_23h = len(posts_photo_23h)

print(f"Nombre de posts de type 'Photo' publiés à 23h : {nombre_posts_photo_23h}")
```

Il y a ici un manque de pertinence avec le taux moyen le plus élevé à 23h, soutenu seulement par une publication. Il faut rappeler que le dataset contient 500 lignes uniques, ce qui est peu pour avoir des insights solides.

## Filtrer les posts Photo

Analyse plus poussée sur les posts de type 'Photo' qui représentent la majorité des publications et le taux d'engagement est le plus élevé.

```python
# Création d'un dataframe filtré sur les posts 'Photos'.
df_photos = df[df['Type'] == 'Photo']
display(df_photos.head())
```

```python
# Calcul du taux d'engagement moyen pour les posts 'Photo' pour chaque jour de la semaine et identifier les jours avec le plus haut engagement.

engagement_photo_par_jour = df_photos.groupby('Post Weekday')['Engagement Rate'].mean().sort_values(ascending=False).round(2)

print("\nTaux d'engagement moyen des posts 'Photo' par jour de la semaine :\n")
display(engagement_photo_par_jour)

jour_plus_engageant_photo = engagement_photo_par_jour.index[0]
taux_plus_engageant_photo = engagement_photo_par_jour.iloc[0]

print(f"\nLe jour de la semaine avec le taux d'engagement le plus élevé pour les posts 'Photo' est le jour {jour_plus_engageant_photo} avec un taux moyen de {taux_plus_engageant_photo:.2f}%")
```

```python
# Calcul du taux d'engagement moyen pour les posts 'Photo' pour chaque heure de la journée et identifier les heures avec le plus haut engagement.

engagement_photo_par_heure = df_photos.groupby('Post Hour')['Engagement Rate'].mean().sort_values(ascending=False).round(2)

print("\nTaux d'engagement moyen des posts 'Photo' par heure de publication :\n")
display(engagement_photo_par_heure)

heure_plus_engageante_photo = engagement_photo_par_heure.index[0]
taux_plus_engageant_photo_heure = engagement_photo_par_heure.iloc[0]

print(f"\nL'heure de publication qui génère le plus d'engagement pour les posts 'Photo' est {heure_plus_engageante_photo}h avec un taux moyen de {taux_plus_engageant_photo_heure:.2f}%.")
```

Même taux que précédement dont la pertinence est faible.

## Insights supplémentaires

```python
# Impact des posts payants (Paid) sur le taux d'engagement
engagement_par_paid = df.groupby('Paid')['Engagement Rate'].mean().sort_values(ascending=False).round(2)
print("\nTaux d'engagement moyen pour les posts payants (sponsorisés) vs non-payants :\n")
display(engagement_par_paid)


taux_paid = engagement_par_paid.iloc[0]
print(f"\nLes posts non sponsorisés ont le taux d'engagement moyen le plus élevé ({taux_paid:.2f}%).")
```

```python
# Calcul du taux d'engagement moyen par Mois de publication
engagement_par_mois = df.groupby('Post Month')['Engagement Rate'].mean().sort_values(ascending=False).round(2)
print("\nTaux d'engagement moyen par Mois de publication :\n")
display(engagement_par_mois)

mois_plus_engageant = engagement_par_mois.index[0]
taux_engageant_mois = engagement_par_mois.iloc[0]
print(f"\nLe mois de publication qui génère le plus d'engagement est le Mois {mois_plus_engageant} avec un taux moyen de {taux_engageant_mois:.2f}%.")
```

## Synthèse des Insights Clés

Voici un résumé des principales observations et recommandations de l'analyse du dataset :

1.  **Performance des Posts 'Photo' et Créneaux Optimaux :**
    *   Les **posts de type 'Photo'** sont le type de contenu le plus engageant en moyenne (taux d'engagement moyen de 2.61%).
    *   Pour les photos, le **Vendredi (Jour 5)** génère le taux d'engagement moyen le plus élevé (2.97%).
    *   L'**heure de publication optimale est 23h**, avec un taux d'engagement moyen de 4.78% pour les photos.
    *   *Insight :* Privilégier la publication de photos le Vendredi soir pour maximiser l'engagement.

2.  **Impact des Posts Sponsorisés (Paid) :**
    *   Les posts **non sponsorisés (Paid = 0)** montrent un taux d'engagement moyen légèrement supérieur (2.48%) aux posts sponsorisés (Paid = 1) qui sont à 2.23%.
    *   *Insight :* Il est intéressant de noter que l'engagement organique peut être plus fort, ou que les campagnes payantes attirent une audience plus large mais potentiellement moins réceptive.

3.  **Taux d'Engagement par Mois :**
    *   Le **Mois de Septembre (Mois 9)** se distingue comme le mois le plus performant en termes d'engagement moyen, avec un taux de 4.17%.
    *   *Insight :* Considérer une attention particulière et peut-être une augmentation de l'activité de publication en Septembre. (Impossible de savoir si c'est saisonnier car le dataset est sur une durée d'un an)
